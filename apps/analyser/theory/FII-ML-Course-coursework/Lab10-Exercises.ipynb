# Exercise 1
Agglomerative clustering on a 2d dataset

Considering the points (-4, -2), (-3, -2), (-2, -2), (-1, -2), (1, -1), (1, 1), (2, 3), (3, 2), (3, 4), (4, 3):
1. create a scatter plot using `pyplot`;
1. create the dendrogram using `AgglomerativeClustering` with single-linkage and then color the scatter plot using the best 4 clusters;
1. create the dendrogram using `AgglomerativeClustering` with complete-linkage and then color the scatter plot using the best 4 clusters;
1. what is the difference in behaviour between the two types of linkage? What shapes do they tend to give to the clusters?

```python
"""
# Exercise 1
Agglomerative clustering on a 2d dataset

Considering the points (-4, -2), (-3, -2), (-2, -2), (-1, -2), (1, -1), (1, 1), (2, 3), (3, 2), (3, 4), (4, 3):
1. create a scatter plot using `pyplot`;
2. create the dendrogram using `AgglomerativeClustering` with single-linkage and then color the scatter plot using the best 4 clusters;
3. create the dendrogram using `AgglomerativeClustering` with complete-linkage and then color the scatter plot using the best 4 clusters;
4. what is the difference in behaviour between the two types of linkage? What shapes do they tend to give to the clusters?
"""


# solution here:
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

points = np.array([[-4, -2], [-3, -2], [-2, -2], [-1, -2], [1, -1], [1, 1], [2, 3], [3, 2], [3, 4], [4, 3]])

# 1. create a scatter plot using `pyplot`;
plt.scatter(points[:, 0], points[:, 1])
plt.show()

# 2. create the dendrogram using `AgglomerativeClustering` with single-linkage and then color the scatter plot using the best 4 clusters;
model = AgglomerativeClustering(n_clusters=4, linkage='single')
model.fit(points)
plt.scatter(points[:, 0], points[:, 1], c=model.labels_)
plt.show()

# 3. create the dendrogram using `AgglomerativeClustering` with complete-linkage and then color the scatter plot using the best 4 clusters;
model = AgglomerativeClustering(n_clusters=4, linkage='complete')
model.fit(points)
plt.scatter(points[:, 0], points[:, 1], c=model.labels_)
plt.show()

# 4. what is the difference in behaviour between the two types of linkage? What shapes do they tend to give to the clusters?
# The single-linkage algorithm tends to create long, thin clusters, while the complete-linkage algorithm tends to create
# more rounded clusters. This is because the single-linkage algorithm tends to merge clusters that are close to each other
# in at least one dimension, while the complete-linkage algorithm tends to merge clusters that are close to each other in
# all dimensions.
```

# Exercise 2
"Natural" clusters

Given the dataset {0, 4, 5, 20, 25, 39, 43, 44}:
1. find the natural clusters using agglomerative clusters with single-linkage and plot clusters using a scatter plot;
1. find the natural clusters using agglomerative clusters with average-linkage and plot clusters using a scatter plot.

```python
"""
# Exercise 2
"Natural" clusters

Given the dataset {0, 4, 5, 20, 25, 39, 43, 44}:
1. find the natural clusters using agglomerative clusters with single-linkage and plot clusters using a scatter plot;
2. find the natural clusters using agglomerative clusters with average-linkage and plot clusters using a scatter plot.
"""

# solution here:

# 1. find the natural clusters using agglomerative clusters with single-linkage and plot clusters using a scatter plot;
points = np.array([0, 4, 5, 20, 25, 39, 43, 44]).reshape(-1, 1)
model = AgglomerativeClustering(n_clusters=2, linkage='single')
model.fit(points)
plt.scatter(points, np.zeros_like(points), c=model.labels_)
plt.show()

# 2. find the natural clusters using agglomerative clusters with average-linkage and plot clusters using a scatter plot.
model = AgglomerativeClustering(n_clusters=2, linkage='average')
model.fit(points)
plt.scatter(points, np.zeros_like(points), c=model.labels_)
plt.show()
```

# Exercise 3

For each the following two datasets `d1` and `d2`:
1. plot the points using `pyplot` and highlight (by using different colours for the points) the 2 clusters found by agglomerative clustering using single linkage;
1. plot the points using `pyplot` and highlight (by using different colours for the points) the 2 clusters found by agglomerative clustering using average linkage.

```python
"""
# Exercise 3

For each the following two datasets `d1` and `d2`:
1. plot the points using `pyplot` and highlight (by using different colours for the points) the 2 clusters found by agglomerative clustering using single linkage;
2. plot the points using `pyplot` and highlight (by using different colours for the points) the 2 clusters found by agglomerative clustering using average linkage.
"""

from sklearn import datasets as ds
import pandas as pd

np.random.seed(0)
X1, _ = ds.make_circles(n_samples=1500, factor=.5,  noise=.05)
X2, _ = ds.make_blobs(n_samples=1500,
                      cluster_std=[1.0, 2.5, 0.5],
                      random_state=170)

d1 = pd.DataFrame(X1, columns=['X1', 'X2'])
d2 = pd.DataFrame(X2, columns=['X1', 'X2'])

# solution here:

# 1. plot the points using `pyplot` and highlight (by using different colours for the points) the 2 clusters found by agglomerative clustering using single linkage;
model = AgglomerativeClustering(n_clusters=2, linkage='single')
model.fit(d1)

plt.scatter(d1['X1'], d1['X2'], c=model.labels_)
plt.show()

model = AgglomerativeClustering(n_clusters=2, linkage='single')
model.fit(d2)

plt.scatter(d2['X1'], d2['X2'], c=model.labels_)
plt.show()

# 2. plot the points using `pyplot` and highlight (by using different colours for the points) the 2 clusters found by agglomerative clustering using average linkage.
model = AgglomerativeClustering(n_clusters=2, linkage='average')
model.fit(d1)

plt.scatter(d1['X1'], d1['X2'], c=model.labels_)
plt.show()

model = AgglomerativeClustering(n_clusters=2, linkage='average')
model.fit(d2)


plt.scatter(d2['X1'], d2['X2'], c=model.labels_)
plt.show()

```
